```python
import nibabel
import os
import cv2
import matplotlib.pyplot as plt
import random
```

```python
path = '/kaggle/input/brats2023-ssa-training-dataset/ASNR-MICCAI-BraTS2023-SSA-Challenge-TrainingData_V2/BraTS-SSA-00007-000/'
t1_path = os.path.join(path, 'BraTS-SSA-00007-000-t1n.nii')
t2_path = os.path.join(path, 'BraTS-SSA-00007-000-t2w.nii')
seg_path = os.path.join(path, 'BraTS-SSA-00007-000-seg.nii')
```

```python
t1 = nibabel.load(t1_path)
t1_data = t1.get_fdata()

t2 = nibabel.load(t2_path)
t2_data = t2.get_fdata()

seg = nibabel.load(seg_path)
seg_data = seg.get_fdata()

t1.shape
```

```python
random_value = random.randrange(1, 155)

t1_slice = t1_data[:, :, random_value]
t2_slice = t2_data[:, :, random_value]
seg_slice = seg_data[:, :, random_value]

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(t1_slice, cmap='gray')
axs[0].set_title("T1 Image")

axs[1].imshow(t2_slice, cmap='gray') 
axs[1].set_title("T2 Image")

axs[2].imshow(seg_slice, cmap='gray')  
axs[2].set_title("Seg Image")

plt.show()
```
